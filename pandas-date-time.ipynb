<h1>why separate objects to handle data and time when python already has datetime library <b>
<i>syntax wise datetime is very convenient
<b>but the performance takes a hit working with huge data .lists vs numpy array
<b>the weakness of python datetime formate inspired numpy team to add a set  of native  series data type to numpy
<b> the datetime64 dtype encodes data as sixtyfourebit intgrs,and thus allows array  of date to be represented very compactly.</h1>

```python
import numpy as np
date=np.array("2025-07-23",dtype=np.datetime64)
```

```python
date
type(date)
```

```python
date.arange(12)
date.np.arange(12)
```

```python
date+np.arange(12)
```

<h1> because of the uniform type in numpy  datetime64 arrays thois type of operation can be accomplised much more quickely than if we were  working directly with pythons datetime objects, espicellay as array get large</h1>
